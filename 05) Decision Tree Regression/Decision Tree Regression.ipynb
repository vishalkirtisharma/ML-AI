```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
df = pd.read_csv('Position_Salaries.csv')
```

```python
X = df.drop(['Salary','Position'],axis=1).values
y = df['Salary'].values
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
regressor = DecisionTreeRegressor(random_state=101)
```

```python
regressor.fit(X,y)
```

```python
regressor.predict([[6.5]])
```

```python
X_grid = np.arange(min(X), max(X), 0.1)
X_grid = X_grid.reshape((len(X_grid), 1))

plt.scatter(X, y, color='red')  # Actual data points
plt.plot(X_grid, regressor.predict(X_grid), color='blue')  # Prediction line
plt.title('Decision Tree Result')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```
