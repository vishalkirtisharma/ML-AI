```python
plt.scatter(x=df['Salary'],y=df['Level'])
```

```python
X =df['Level'].values.reshape(-1,1)
y =df['Salary'].values.reshape(-1,1)
```

```python
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
```

```python
lin_reg.fit(X,y)
```

```python
from sklearn.preprocessing import PolynomialFeatures

poly_reg = PolynomialFeatures(degree=4)
X_poly = poly_reg.fit_transform(X)
```

```python
lin_reg_2 = LinearRegression()
```

```python
lin_reg_2.fit(X_poly,y)
```

```python
plt.scatter(X,y,color='red')
plt.plot(X,lin_reg.predict(X))

plt.title('Lineaer Regression')
plt.xlabel('Position Level')
plt.ylabel('Salary')

plt.show()
```

```python
plt.scatter(X,y,color='red')
plt.plot(X,lin_reg_2.predict(poly_reg.transform(X)))

plt.title('Lineaer Regression')
plt.xlabel('Position Level')
plt.ylabel('Salary')

plt.show()
```

```python
print('With Linear regression')
lin_reg.predict([[6.5]])
```

```python
print('With Polynomial  regression')
lin_reg_2.predict(poly_reg.transform([[6.5]]))
```
